```python
import json

class Paciente:
    def __init__(self, nome: str, nascimento: str, cpf: str, sexo: str, altura: int, peso: int, imc: float, plano_de_saude: str):
        self.nome: str = nome
        self.nascimento: str = nascimento
        self.cpf: str = cpf
        self.sexo: str = sexo
        self.altura: int = altura
        self.peso: int = peso
        self.imc: float = imc
        self.plano_de_saude: str = plano_de_saude

pacientes = []

def calcular_imc(peso: int, altura: int) -> float:
    altura_metros = altura / 100
    imc = peso / (altura_metros ** 2)
    return round(imc, 2)

def cadastrar_paciente():
    nome = input("Digite o nome: ")
    nascimento = input("Digite a data de nascimento: ")
    cpf = input("Digite o CPF: ")
    sexo = input("Digite o sexo: ")
    altura = input("Digite a altura em cm: ")
    peso = input("Digite o peso em kg: ")

    try:
        altura = int(altura)
        peso = int(peso)
    except ValueError:
        print("Por favor, insira valores válidos para altura e peso.")
        return
    
    possui_plano = input("Possui plano de saúde? (s/n): ").strip().lower()
    if possui_plano == 's':
        plano_de_saude = input("Digite o nome do plano de saúde: ")
    else:
        plano_de_saude = "Particular"
    
    imc = calcular_imc(peso, altura)
    
    novo_paciente = Paciente(nome, nascimento, cpf, sexo, altura, peso, imc, plano_de_saude)
    pacientes.append(novo_paciente)
    print(f"Paciente {nome} cadastrado com sucesso!")

def remover_paciente():
    cpf = input("Digite o CPF do paciente a ser removido: ")
    global pacientes
    pacientes = [paciente for paciente in pacientes if paciente.cpf != cpf]
    print(f"Paciente com CPF {cpf} removido com sucesso!")

def alterar_cadastro_paciente():
    cpf = input("Digite o CPF do paciente a ser alterado: ")
    for paciente in pacientes:
        if paciente.cpf == cpf:
            print("Deixe o campo em branco se não quiser alterar o valor.")
            paciente.nome = input(f"Nome ({paciente.nome}): ") or paciente.nome
            paciente.nascimento = input(f"Ano de Nascimento ({paciente.nascimento}): ") or paciente.nascimento
            paciente.sexo = input(f"Sexo ({paciente.sexo}): ") or paciente.sexo
            altura = input(f"Altura em cm ({paciente.altura}): ") or paciente.altura
            peso = input(f"Peso em kg ({paciente.peso}): ") or paciente.peso
            try:
                altura = int(altura)
                peso = int(peso)
            except ValueError:
                print("Por favor, insira valores válidos para altura e peso.")
                return
            paciente.altura = altura
            paciente.peso = peso
            paciente.imc = calcular_imc(peso, altura)
            possui_plano = input(f"Possui plano de saúde ({paciente.plano_de_saude})? (s/n): ").strip().lower()
            if possui_plano == 's':
                paciente.plano_de_saude = input(f"Nome do plano de saúde ({paciente.plano_de_saude}): ") or paciente.plano_de_saude
            else:
                paciente.plano_de_saude = "Particular"
            print(f"Cadastro do paciente com CPF {cpf} atualizado com sucesso!")
            return
    print(f"Paciente com CPF {cpf} não encontrado.")

def consultar_cadastro_paciente():
    cpf = input("Digite o CPF do paciente a ser consultado: ")
    for paciente in pacientes:
        if paciente.cpf == cpf:
            print(vars(paciente))
            return
    print(f"Paciente com CPF {cpf} não encontrado.")

def listar_pacientes():
    for paciente in pacientes:
        print(vars(paciente))

def salvar_dados_json():
    dados_para_salvar = []
    for paciente in pacientes:
        dados_paciente = {
            'nome': paciente.nome,
            'nascimento': paciente.nascimento,
            'cpf': paciente.cpf,
            'sexo': paciente.sexo,
            'altura': paciente.altura,
            'peso': paciente.peso,
            'imc': paciente.imc,
            'plano_de_saude': paciente.plano_de_saude
        }
        dados_para_salvar.append(dados_paciente)
    
    with open('dados.json', 'w') as arquivo_json:
        json.dump(dados_para_salvar, arquivo_json, indent=4)

def carregar_dados_json():
    global pacientes
    pacientes = []
    try:
        with open('dados.json', 'r') as arquivo_json:
            dados = json.load(arquivo_json)
            for dados_paciente in dados:
                paciente = Paciente(
                    dados_paciente['nome'],
                    dados_paciente['nascimento'],
                    dados_paciente['cpf'],
                    dados_paciente['sexo'],
                    dados_paciente['altura'],
                    dados_paciente['peso'],
                    dados_paciente['imc'],
                    dados_paciente['plano_de_saude']
                )
                pacientes.append(paciente)
    except FileNotFoundError:
        print("Arquivo de dados não encontrado.")
    except json.JSONDecodeError:
        print("Erro ao decodificar o arquivo de dados.")

# Menu para interagir com o usuário
def menu():
    while True:
        print("\nMenu:")
        print("1. Cadastrar paciente")
        print("2. Remover paciente")
        print("3. Alterar cadastro do paciente")
        print("4. Consultar cadastro do paciente")
        print("5. Listar pacientes")
        print("6. Salvar dados")
        print("7. Carregar dados")
        print("8. Sair")
        opcao = input("Escolha uma opção: ")
        
        if opcao == '1':
            cadastrar_paciente()
        elif opcao == '2':
            remover_paciente()
        elif opcao == '3':
            alterar_cadastro_paciente()
        elif opcao == '4':
            consultar_cadastro_paciente()
        elif opcao == '5':
            listar_pacientes()
        elif opcao == '6':
            salvar_dados_json()
        elif opcao == '7':
            carregar_dados_json()
        elif opcao == '8':
            break
        else:
            print("Opção inválida, tente novamente.")

# Executa o menu
menu()
```
